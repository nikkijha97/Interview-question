```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'name':['A','B','C','D'],
                  'salary':[10,20,78,11]})
df
```

```python
## add a new column 'status' according to the  salary levels <20 low, 20-50 mid, >50 high
con=[
    (df['salary']<20),
     (df['salary']>=20) & (df['salary']<50),
      (df['salary']>50)
]

r=['low','mid','high']

df['status']=np.select(con, r)
print(df)
```

```python
## change the  row name according to the status
status_inx=df.set_index('status')
df=status_inx
df
```
