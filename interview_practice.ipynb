```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Parameters for data generation
num_records = 1000
num_stores = 5
num_products = 20
num_customers = 100

# Generate sample data
store_ids = [f'ST{str(i).zfill(2)}' for i in range(1, num_stores + 1)]
product_ids = [f'PR{str(i).zfill(3)}' for i in range(1, num_products + 1)]
customer_ids = [f'CUST{str(i).zfill(3)}' for i in range(1, num_customers + 1)]

# Generate random dates within a specific range
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = (end_date - start_date).days
# Create the DataFrame
data = {
    'Date': [(start_date + timedelta(days=random.randint(0, date_range))).strftime('%Y-%m-%d') for _ in range(num_records)],
    'Store_ID': [random.choice(store_ids) for _ in range(num_records)],
    'Product_ID': [random.choice(product_ids) for _ in range(num_records)],
    'Quantity': [random.randint(1, 10) for _ in range(num_records)],
    'Unit_Price': [round(random.uniform(5.0, 100.0), 2) for _ in range(num_records)],
    'Customer_ID': [random.choice(customer_ids) for _ in range(num_records)]
}

# Create DataFrame
sales_data = pd.DataFrame(data)

sales_data
```

```python
sales_data['top_prod']=sales_data['Quantity']*sales_data['Unit_Price']
sales_prod_data=sales_data.groupby(['Store_ID','Product_ID'])['top_prod'].sum().reset_index()
top_sell=sales_prod_data.loc[sales_prod_data.groupby('Store_ID')['top_prod'].idxmax()]
top_sell
```

```python
import pandas as pd

# Sample Products DataFrame
products = pd.DataFrame({
    'ProductID': [1, 2, 3, 4],
    'ProductName': ['Laptop', 'Phone', 'Chair', 'Table'],
    'Price': [800, 500, 150, 300],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture']
})

# Sample Sales DataFrame
sales = pd.DataFrame({
    'SaleID': [1, 2, 3, 4],
    'ProductID': [1, 2, 3, 4],
    'QuantitySold': [5, 7, 10, 2]
})

```

```python
# Merge the Products and Sales DataFrames
merged = pd.merge(products, sales, on='ProductID')

```

```python
# Filter for Electronics Category
electronics = merged[merged['Category'] == 'Electronics']

```

```python
merged
```

```python
# Calculate Revenue (Price * QuantitySold)
electronics['Revenue'] = electronics['Price'] * electronics['QuantitySold']
```

```python
products
```

```python
# Filter for Products with Revenue > 3000
filtered_products = electronics[electronics['Revenue'] > 3000]

# Fetch Product Names
result = filtered_products['ProductName'].tolist()

print("Electronics products with revenue > 3000:", result)
```
