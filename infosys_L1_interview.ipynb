```python
import numpy as np
import pandas as pd
```

```python
# Creating a sample dataset
data = {
    'Customer_ID': [101, 102, 103, 101, 104, 105, 102, 103, 106, 107],
    'Transaction_Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Product_Category': ['Electronics', 'Clothing', 'Grocery', 'Electronics', 'Grocery', 'Clothing', 'Electronics', 'Grocery', 'Clothing', 'Electronics'],
    'Amount': [1200, 800, 300, 1500, 250, 600, 1300, 400, 700, 1600],
    'Discount': [100, np.nan, 20, 150, 10, np.nan, 120, 30, np.nan, 180],
    'Payment_Method': ['Credit Card', 'UPI', 'Cash', 'Debit Card', 'Credit Card', 'UPI', 'Debit Card', 'Cash', 'UPI', 'Credit Card'],
    'Store_Location': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Chennai', 'Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Chennai']
}
```

```python
df=pd.DataFrame(data)
df
```
